# Remove Duplicates

```python
print("\n--- Before Removing Duplicates ---")
print(df.shape)

df = df.drop_duplicates()

print("After Removing Duplicates:", df.shape)
```

# Handle Missing Values

```python
# Drop columns with more than 60% missing values
threshold = 0.6
df = df.loc[:, df.isnull().mean() < threshold]

# Fill missing categorical with mode and numeric with median
for col in df.columns:
    if df[col].dtype == "object":
        df[col] = df[col].fillna(df[col].mode()[0])
    else:
        df[col] = df[col].fillna(df[col].median())

print("\nMissing values left:", df.isnull().sum().sum())
```

# Encode Categorical Variables

```python
categorical_cols = df.select_dtypes(include="object").columns
encoder = LabelEncoder()

for col in categorical_cols:
    try:
        df[col] = encoder.fit_transform(df[col].astype(str))
    except Exception as e:
        print(f"Could not encode {col}: {e}")

print("\n--- Encoded Dataset Preview ---")
print(df.head())
```

# Generate Insights

```python
def generate_insights(data):
    insights = {}

    # Programming languages (Q9 parts)
    prog_cols = [col for col in data.columns if "q9" in col.lower()]
    if prog_cols:
        top_langs = data[prog_cols].apply(pd.Series.value_counts).sum(axis=1).sort_values(ascending=False).head(5)
        insights["Top Programming Languages"] = top_langs

    # IDEs
    if "q13" in data.columns:
        insights["Top IDEs"] = data["q13"].value_counts().head(5)

    # Education
    for edu_col in ["q4", "q5"]:
        if edu_col in data.columns:
            insights["Top Education Levels"] = data[edu_col].value_counts().head(5)
            break

    # Job titles
    for job_col in ["q6", "q7_part_1"]:
        if job_col in data.columns:
            insights["Top Job Titles"] = data[job_col].value_counts().head(5)
            break

    # Salary
    if "q25" in data.columns:
        insights["Top Salary Ranges"] = data["q25"].value_counts().head(5)

    return insights
```

```python
yearly_trends = {"Programming Languages": {}, "IDEs": {}, "Education": {}, "Job Titles": {}, "Salaries": {}}

if "year" in df.columns:
    years = sorted(df["year"].unique())
    for y in years:
        print(f"\n========== Year {y} ==========")
        df_year = df[df["year"] == y].copy()
        df_year = clean_data(df_year)
        
        insights = generate_insights(df_year)
        
        for k, v in insights.items():
            print(f"\n{k}:\n{v}")
        
        plot_dashboard(insights, year=y)
        
        # Store top 1 for trend comparison
        if "Top Programming Languages" in insights:
            yearly_trends["Programming Languages"][y] = insights["Top Programming Languages"].idxmax()
        if "Top IDEs" in insights:
            yearly_trends["IDEs"][y] = insights["Top IDEs"].idxmax()
        if "Top Education Levels" in insights:
            yearly_trends["Education"][y] = insights["Top Education Levels"].idxmax()
        if "Top Job Titles" in insights:
            yearly_trends["Job Titles"][y] = insights["Top Job Titles"].idxmax()
        if "Top Salary Ranges" in insights:
            yearly_trends["Salaries"][y] = insights["Top Salary Ranges"].idxmax()
```
